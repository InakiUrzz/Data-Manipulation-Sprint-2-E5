### Columna de velocidad media

```python
df_n['velocidad media'] = df_n['Distance']/df_n['AirTime']
df_n.head()
```

### Columna de arribos tardios

```python
def arribos_tardios(x):
    if x > 0:
        return "Arribo Tarde"
    else: 
        return "Sin Demora"
df_n['arribos tardios'] = df_n['ArrDelay'].apply(arribos_tardios)
df.head()
```

### Vuelos con mayor atrasos acumulado

```python
prom_carry_d = df_n['CarrierDelay'].mean()
```

```python
carry_delay = df_n[df_n['CarrierDelay']> prom_carry_d]
carry_delay.head()
```

### Vuelos mas largos

```python
prom_airtime = df_n['AirTime'].mean()
```

```python
vuelos_mas_largos = df_n[df_n['AirTime']>prom_airtime]
vuelos_mas_largos.head()
```

### Vuelos con mÃ¡s atrasados

```python
prom_demora = df_n['ArrDelay'].mean()
prom_demora
```

```python
vuelos_mas_atrasados = df_n[df_n['ArrDelay'] > prom_demora]
vuelos_mas_atrasados.head(9)
```
